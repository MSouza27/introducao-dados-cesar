```python
df = df[df['country'] != '']
df_country = df.groupby('country')['duration'].count().reset_index().rename(columns={'duration': 'total'}).sort_values(by='total', ascending=False)
df_country = df_country[df_country['total'] > 100]
df_country
```

```python
df_country.plot(kind='bar',
                x='country',
                y='total')
```

```python
df_type = df.groupby('type')['duration'].count().reset_index().rename(columns={'duration': 'total'}).sort_values(by='total', ascending=False)
df_type
```

```python
df_type.plot.bar()
```

```python
df_year = df.groupby('release_year')['show_id'].count().reset_index().rename(columns={'show_id': 'total'}).sort_values(by='release_year', ascending=True)
df_year = df_year[df_year['release_year'] >= 2016]
df_year
```

```python
df_year.plot(kind='bar',
             x='release_year',
             y='total')
```
