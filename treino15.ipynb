```python
df_model = df.groupby(['Model', 'Region'])['Avg_Price_USD'].sum().reset_index().sort_values(by='Avg_Price_USD', ascending=False)
df_model
```

```python
df_grafico_model = px.bar(
    df_model,
    x='Model',
    y='Avg_Price_USD',
    color='Region',
    text_auto=True
)
df_grafico_model.show()
```

```python
def ajuste_regiao(regiao):
    if regiao == 'Europe':
        return 'Europa'
    elif regiao == 'North America':
        return 'America do Norte'
    else:
        return 'Asia'

df['nova_region'] = df['Region'].apply(ajuste_regiao)

df.head()
```
