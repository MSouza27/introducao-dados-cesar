```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../archive/precos_carros_brasil - precos_carros_brasil.csv')
df.head()
```

```python
df = df.drop_duplicates(keep='last')
```

```python
df.duplicated().sum()
```

```python
df_mes_atualizado = {
    'January': 'Janeiro',
    'February': 'Fevereiro',
    'March': 'Março',
    'April': 'Abril',
    'May': 'Maio',
    'June': 'Junho',
    'July': 'Julho',
    'September': 'Setembro',
    'August': 'Agosto',
    'October': 'Outubro',
    'November': 'Novembro',
    'December': 'Dezembro'
}
```

```python
df['month_of_reference'] = df['month_of_reference'].map(df_mes_atualizado)
```

```python
df = df[df['year_of_reference'] == 2021]
df_vendas = df.groupby('month_of_reference')['authentication'].count().reset_index().rename(columns={'authentication': 'total'}).sort_values(by='total')
df_vendas
```

```python
df_bar = px.bar(
    df_vendas,
    x='month_of_reference',
    y='total',
    title='Acompanhamento de Vendas',
    color_discrete_sequence=px.colors.qualitative.Bold,
    labels={
        'month_of_reference' : 'Mês de Referência',
        'total' : 'Quantidade Vendido'
    }
)
df_bar.update_traces(marker_color='blue')
df_bar.show()
```
