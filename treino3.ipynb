```python
df = df[df['release_year'] >= 2018]
df_ano = df.groupby('release_year')['show_id'].count().reset_index().rename(columns={'show_id': 'total'}).sort_values(by='release_year', ascending=True)
df_ano
```

```python
df_grafico = px.line(
    df_ano,
    x='release_year',
    y='total',
    markers=True,
    title='Filmes Por Ano',
    labels={
        'release_year': 'Ano',
        'total': 'Vendas'
    }
)
df_grafico.show()
```

```python
df = df[df['release_year'] == 2021]
df_country = df.groupby(['country', 'type'])['show_id'].count().reset_index().rename(columns={'show_id': 'total'}).sort_values(by='total', ascending=False)
df_country = df_country[df_country['total'] >= 13]
df_country
```

```python
df_dados = px.bar(
    df_country,
    x='country',
    y='total',
    color='type',  # coluna correta
    color_discrete_sequence=px.colors.qualitative.Bold,
    title='Filmes por tipo e país',
    labels={
        'country': 'País',
        'total': 'Quantidade de Filmes',
        'type': 'Tipo de Filme'
    },
    barmode='stack'  # ou 'group' / 'overlay' conforme o desejado
)

df_dados.show()
```
