```python
df = df[df['Weight (kg)'] >= 73]
df_dados_persona = df.groupby(['Gender', 'Workout_Type'])['Max_BPM'].count().reset_index().rename(columns={'Max_BPM': 'total'}).sort_values(by='total', ascending=False)
df_dados_persona
```

```python
df_grafico = px.bar(
    df_dados_persona,
    x='Workout_Type',
    y='total',
    color='Gender',
    title='Dados Perfil Genero',
    color_discrete_sequence=px.colors.qualitative.Bold,
    labels={
        'Workout_Type' : 'Atividade',
    }
)
#df_grafico.update_traces(marker_color='green')
df_grafico.show()
```
