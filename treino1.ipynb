```python
df_age = df.groupby('Gender')['Age'].count().reset_index().rename(columns={'Age' : 'total'}).sort_values(by='total', ascending=False)
df_age
```

```python
df_age.plot(kind='bar',
            x='Gender',
            y='total')
```

```python
df = df[df['Age'] >= 55]
df_max_bpm = df.groupby(['Gender', 'Max_BPM'])['Age'].mean().reset_index().rename(columns={'Age': 'total'}).sort_values(by='total', ascending=False)
df_max_bpm
```

```python
df.value_counts('Workout_Type')
```

```python
df = df[df['Max_BPM'] >= 162]
df_bpm = df.groupby('Gender')['Age'].count().reset_index().rename(columns={'Age': 'total'}).sort_values(by='total', ascending=False)
df_bpm
```
