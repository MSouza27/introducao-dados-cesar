```python
df_location = df.groupby('location')['student_id'].count().reset_index().rename(columns={'student_id': 'total_estudantes'}).sort_values(by='total_estudantes', ascending=False)
df_location
```

```python
df_graph = px.bar(
    df_location,
    x='location',
    y='total_estudantes'
)
df_graph.show()
```

```python
df_disciplinas = ['sql_marks', 'excel_marks', 'python_marks', 'power_bi_marks', 'english_marks']
df_media_nota = df.groupby('location')[df_disciplinas].mean().reset_index().round(2)
df_media_nota
```

```python
df_grafico_notas = px.bar(
    df_media_nota,
    x='location',
    y=df_disciplinas,
    color_discrete_sequence=px.colors.qualitative.Bold,
    barmode='stack',
    title='Analise média das notas',
    labels={
        'location':'Localização',
    }
)

df_grafico_notas.show()
```

```python
perfil = {
    18: 'Jovem',
    19: 'Jovem',
    20: 'Jovem',
    21: 'Adulto Jovem',
    22: 'Adulto Jovem',
    23: 'Adulto Jovem',
    24: 'Adulto Jovem',
    25: 'Adulto'
}

df['perfil'] = df['age'].map(perfil)
df_perfil = (df.groupby(['location', 'perfil'])['age']
             .count()
             .reset_index()
             .rename(columns={'age': 'total'})
             .sort_values(by='total', ascending=False))
df_perfil
```

```python
df_grafico_perfil = px.bar(
    df_perfil,
    x='location',
    y='total',
    color='perfil'
)
df_grafico_perfil.show()
```

```python
def classificar_nota(nota):
    if nota < 50:
        return 'Reprovado'
    elif nota < 70:
        return 'Em Recuperação'
    elif nota < 90:
        return 'Aprovado'
    else:
        return 'Excelente'

df['media_geral'] = df[['sql_marks', 'excel_marks', 'python_marks', 'power_bi_marks', 'english_marks']].mean(axis=1)
df['status_nota'] = df['media_geral'].apply(classificar_nota)

df_analise_nota = (
    df.groupby(['location', 'status_nota'])['student_id']
    .count()
    .reset_index()
    .rename(columns={'student_id': 'total'})
    .sort_values(by='total', ascending=False)
)

df_analise_nota

```

```python
df_dados_grafico = px.bar(
    df_analise_nota,
    x='location',
    y='total',
    color='status_nota'
)
df_dados_grafico.show()
```
