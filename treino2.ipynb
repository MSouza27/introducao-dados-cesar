```python
df = df[df['week'] >= 50]
df_week = df.groupby('week')['staff_id'].count().reset_index().rename(columns={'staff_id' : 'total'}).sort_values(by='week', ascending=True)
df_week
```

```python
df = df[df['role'] == 'doctor']
df_role = df.groupby(['service', 'present'])['staff_id'].count()
df_role
```
