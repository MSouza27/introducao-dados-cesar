```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df_nobel = pd.read_csv('/Users/magnosouza/PycharmProjects/IntroducaoAnaliseDados/archive/nobel_prizes_1901-2025_cleaned.csv')
df_nobel.head()
```

```python
df_nobel_ano = df_nobel.groupby(['award_year', 'sex'])['prize_amount_adjusted'].sum().reset_index().sort_values(by='prize_amount_adjusted',ascending=False)
df_nobel_ano
```

```python
df_dados_ano = px.bar(
    df_nobel_ano,
    x='award_year',
    y='prize_amount_adjusted',
    color='sex',
    color_discrete_sequence=px.colors.qualitative.Bold,
    barmode='stack',
    title='Analise Nobel Por Sexo x Valor',
    labels= {
        'award_year' : 'Ano Prêmio',
        'prize_amount_adjusted' : 'Valor Prêmio',
        'sex': 'Sexo'
    }
)
df_dados_ano.show()
```

```python
df_nobel_sex = df_nobel.groupby('sex')['prize_amount_adjusted'].sum().reset_index().sort_values(by='prize_amount_adjusted',ascending=False)
df_nobel_sex
```

```python
df_dados_sex = px.bar(
    df_nobel_sex,
    x='prize_amount_adjusted',
    y='sex',
    title='Analise Prêmio por Sexo'
)
df_dados_sex.show()
```

```python
df_nobel_city = df_nobel.groupby(['affiliation_city', 'sex'])['prize_amount_adjusted'].sum().reset_index().sort_values(by='prize_amount_adjusted', ascending=False)
df_nobel_city
```

```python
df_dados_city = px.bar(
    df_nobel_city,
    x='affiliation_city',
    y='prize_amount_adjusted',
    color='sex',
    color_discrete_sequence=px.colors.qualitative.Bold,
    barmode='stack'
)
df_dados_city.show()
```

```python
ano = df_nobel['award_year']
sex = df_nobel['sex']
valor = df_nobel['prize_amount_adjusted']
```

```python
plt.style.use('dark_background')
```

```python
fig = plt.figure(figsize=(12,8))
grid = fig.add_gridspec(2, 2)
```

```python
ax_linha = fig.add_subplot(grid[0, :])
ax_hist1 = fig.add_subplot(grid[1, 0])
ax_hist2 = fig.add_subplot(grid[1, 1])
```

```python
ax_linha.plot(ano, valor,  color='green', linestyle='--', marker='.')
ax_linha.set_title('Valores do Prêmio Por ano')
ax_linha.set_xlabel('Ano')
ax_linha.set_ylabel('Valor Prêmio')
ax_linha.legend()
plt.tight_layout()


ax_hist1.hist(valor, color='orange')


display(fig)
plt.show()
```
