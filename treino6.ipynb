```python
df = df[df['type'] == 'Movie']
df_movie = df.groupby('country')['show_id'].count().reset_index().rename(columns={'show_id': 'total_cidade'}).sort_values(by='total_cidade', ascending=False)
df_movie = df_movie[df_movie['total_cidade'] >= 100]
df_movie
```

```python
df_grafico_bar = px.bar(
    df_movie,
    x='country',
    y='total_cidade',
    title='Analise de Filmes por Cidade',
    color_discrete_sequence=px.colors.qualitative.Bold,
    labels={
        'country': 'cidade',
        'total_cidades': 'contagem'
    }
)

df_grafico_bar.update_traces(marker_color='blue')
df_grafico_bar.show()
```
