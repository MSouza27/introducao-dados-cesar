```python
import pandas as pd
import plotly.express as px
```

```python
df_bmw = pd.read_csv('../archive/BMW sales data (2010-2024) (1).csv')
df_bmw.head()
```

```python
df_bmw.info()
```

```python
df_bmw.isna().any()
```

```python
df_bmw.duplicated().any()
```

```python
df_bmw.describe().round(2)
```

```python
df_bmw_vendas = df_bmw.groupby(['Region', 'Model'])['Sales_Volume'].sum().reset_index().rename(columns={'Sales_Volume': 'total_vendidos'}).sort_values('total_vendidos', ascending=False)
df_bmw_vendas = df_bmw_vendas[df_bmw_vendas['total_vendidos'] >= 327583 ]
df_bmw_vendas
```

```python
df_analytics = px.bar(
    df_bmw_vendas,
    x='Region',
    y='total_vendidos',
    color='Model',
    title='Vendas BMW',
    color_discrete_sequence=px.colors.qualitative.Bold
)
df_analytics.show()
```

```python
df_bmw_ano = df_bmw.groupby('Year')['Sales_Volume'].sum().reset_index().sort_values(by='Year', ascending=True).rename(columns={'Sales_Volume': 'vendas'})
df_bmw_ano
```

```python
df_histograma = px.line(
    df_bmw_ano,
    x='Year',
    y='vendas',
    title='Vendas por Ano',
    labels={
        'Year':'Ano de Venda',
        'vendas': 'Quantidade Vendida'
    },
    markers=True
)
df_histograma.show()
```

```python
df_bmw = df_bmw[df_bmw['Year'] == 2022]
df_bmw_max_vendas = df_bmw.groupby('Model')['Sales_Volume'].sum().reset_index().rename(columns={'Sales_Volume': 'vendas'}).sort_values(by='vendas', ascending=False)
df_bmw_max_vendas.head()
```

```python
df_max_vendas = px.bar(
    df_bmw_max_vendas.head(),
    x='Model',
    y='vendas',
    title='Top 5 Vendas no Ano 2022',
    labels={
        'Model': 'Modelo Vendido',
        'vendas': 'Quantidade Vendida'
    }
)
df_max_vendas.update_traces(marker_color='green')
df_max_vendas.show()
```

```python
df_bmw = df_bmw[df_bmw['Year'] == 2022]
df_bmw_region = df_bmw.groupby('Region')['Sales_Volume'].sum().reset_index().sort_values(by='Sales_Volume',ascending=False)
df_bmw_region
```

```python
df_region = px.bar(
    df_bmw_region,
    x='Region',
    y='Sales_Volume',
    title='Vendas por Regio Ano 2022'
)
df_region.show()
```
