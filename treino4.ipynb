```python
from pickletools import markobject

import pandas as pd
import plotly.express as px
```

```python
df_dados_hospital = pd.read_csv('../archive/patients.csv')
df_dados_hospital.head()
```

```python
df_dados_hospital.isna().sum()
```

```python
df_dados_hospital.duplicated().sum()
```

```python
df_dados_hospital = df_dados_hospital[df_dados_hospital['age'] < 18]
df_menores = df_dados_hospital.groupby('service')['name'].count().reset_index().rename(columns={'name': 'total'}).sort_values(by='total', ascending=False)
df_menores
```

```python
df_grafico = px.bar(
    df_menores,
    x='service',
    y='total',
    color_discrete_sequence=px.colors.qualitative.Bold,
    title='Analise de Atendimento de menores',
    labels={
        'service': 'Tipo de Atendimento',
        'total' : 'Quantidade de Atendimento'
    }

)
df_grafico.update_traces(marker_color='green')
df_grafico.show()
```
