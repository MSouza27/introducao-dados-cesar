```python
#df = df[df['service'] == 'general_medicine']
df_name = df.groupby(['service'])['patient_id'].count().reset_index().rename(columns={'patient_id': 'total_atendimento'}).sort_values(by='total_atendimento', ascending=False)
df = df[df['satisfaction'] >= 80]
df_name
```

```python
df_grafico = px.bar(
    df_name,
    x='service',
    y='total_atendimento',
    color_discrete_sequence=px.colors.qualitative.Bold,
    title='Total de Atendimento avaliados acima de 80',
    labels={
        'service' : 'Especialidade',
        'total_atendimento' : 'Total Atendimento'
    }
)
df_grafico.update_traces(marker_color='green')
df_grafico.show()
```
