```python
df_brand = df.groupby('brand')['model'].count().reset_index().rename(columns={'model': 'total'}).sort_values(by='total', ascending=False)
df_brand
```

```python
df_brand_grafico = px.bar(
    df_brand,
    x='brand',
    y='total',
    text_auto=True,
    title='Dados por Empresa',
    labels={
        'brand': 'Empresa',
        'total': 'Total'
    }
)
df_brand_grafico.show()
```

```python
df['release_month'].duplicated().head(12)
```

```python
meses = {
    'January': 1,
    'February' : 2,
    'March' : 3,
    'April' : 4,
    'May' : 5,
    'June' : 6,
    'July' : 7,
    'August' : 8,
    'September' : 9,
    'October' : 10,
    'November' : 11,
    'December' : 12
}

df['month_index'] = df['release_month'].map(meses)

df_usd = df.groupby(['brand', 'release_month', 'month_index'])['price_usd'].sum().reset_index().sort_values(by='month_index', ascending=True)
df_usd
```

```python
df_usd_grafico = px.bar(
    df_usd,
    x='release_month',
    y='price_usd',
    color='brand',
    text_auto=True,
    color_discrete_sequence=px.colors.qualitative.Bold
)
df_usd_grafico.show()
```

```python
df.head()
```

```python
df_os = df.groupby(['os', 'release_month', 'month_index'])['price_usd'].sum().reset_index().sort_values(by='month_index', ascending=True)
df_os
```

```python
df_os_grafico = px.bar(
    df_os,
    x='release_month',
    y='price_usd',
    color='os',
    color_discrete_sequence=px.colors.qualitative.Bold
)
df_os_grafico.show()
```
